# Figures of GNN Regression Predictions on Experimental Data

This script is to check the exp param kde distributions if we directly apply GNN regressor (i.e., s1) on the exp data without using ABC.

### Simulation Date: 2026-03-15

Comments:
1. gamma range = [1, 1000]
2. alpha range = [0.1, 100]
3. exp data z-score normalized using their own mean&std
4. picklesteps = 5
5. connectivity issue not solved yet

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
def plot_kde(data):
    data = np.array(data)

    # KDE
    kde = gaussian_kde(data)
    x_vals = np.linspace(data.min() - 1, data.max() + 1, 500)
    y_vals = kde(x_vals)

    # Mean
    mean = np.mean(data)

    # Plot
    plt.plot(x_vals, y_vals, label="KDE")
    plt.hist(data, bins=10, density=True, alpha=0.3, label="Histogram")

    # Vertical line for mean
    plt.axvline(mean, linestyle="--", linewidth=2, label=f"Mean = {mean:.2f}")

    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.title("KDE Distribution")
    plt.legend()
    plt.show()

# Example
#numbers = [1, 2, 2, 3, 4, 5, 5, 6]
#plot_kde(numbers)
```
