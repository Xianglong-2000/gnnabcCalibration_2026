```python
import pickle as pk
import torch
import numpy as np
import os
import pandas as pd
```

```python
g_path = "s6_0313_01.pkl"

with open(g_path, "rb") as f:
    data = pk.load(f)

print(type(data))
print(len(data))
print(type(list(data.values())[0]))
```

```python
bad_graphs = []

for i, g in enumerate(data.values()):

    if hasattr(g, "x"):
        if torch.isnan(g.x).any() or torch.isinf(g.x).any():
            print(f"Graph {i} has NaN/Inf in x")
            bad_graphs.append(i)

    if hasattr(g, "y"):
        if torch.isnan(g.y).any() or torch.isinf(g.y).any():
            print(f"Graph {i} has NaN/Inf in y")
            print(g.y)
            bad_graphs.append(i)

    if hasattr(g, "edge_attr") and g.edge_attr is not None:
        if torch.isnan(g.edge_attr).any() or torch.isinf(g.edge_attr).any():
            print(f"Graph {i} has NaN/Inf in edge_attr")
            bad_graphs.append(i)

print("Bad graphs:", bad_graphs)
```

```python
def smart_cast(val):
    try:
        return float(val)
    except ValueError:
        if val == 'True':
            return float(1)
        elif val == 'False':
            return float(0)
        elif val == 'TRUE':
            return float(1)
        elif val == 'FALSE':
            return float(0)
        elif val == 'YFP':
            return float(1)
        else:
            return val  # keep as string if not a number
        
include_columns = ['ImageNumber', #
                       'ObjectNumber', #
                       'id', #
                       'parent_id', #
                       'AreaShape_Area', #
                       'AreaShape_MajorAxisLength', #
                       'AreaShape_Orientation', #
                       'cellAge', #
                       'LifeHistory', #
                       'TrajectoryX', #
                       'TrajectoryY', #
                       'Direction_of_Motion', #
                       'Motion_Alignment_Angle', #
                       'Source_Neighbor_Avg_TrajectoryX', #
                       'Source_Neighbor_Avg_TrajectoryY', #
                       'Division_TimeStep', #
                       'Daughter_Mother_Length_Ratio', #
                       'Total_Daughter_Mother_Length_Ratio', #
                       'Max_Daughter_Mother_Length_Ratio', #
                       'Daughter_Avg_TrajectoryX', #
                       'Daughter_Avg_TrajectoryY', #
                       'Neighbor_Difference_Count', #
                       'Neighbor_Shared_Count', #
                       'Average_Length', #
                       'Velocity', #
                       'Instant_Velocity', #
                       'Average_Instant_Velocity', #
                       'Elongation_Rate', #
                       'Instant_Elongation_Rate', #
                       'startVol', #
                       'targetVol', #
                       'Prev_MajorAxisLength', #
                       'Bacterium_Movement', #
                       'dir_1', 'dir_2', #
                       ]
```
